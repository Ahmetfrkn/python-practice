```python
gorev_list = []

def add_gorev():
    yeni_gorev = input("Ne görev eklemek istiyorsunuz? ")
    gorev_list.append(yeni_gorev)
    print("Görev eklendi!")

def show_gorev():
    if not gorev_list:
        print("Henüz görev yok.")
    else:
        print("\nYapılacaklar Listesi:")
        for i, gorev in enumerate(gorev_list, 1):
            print(f"{i}. {gorev}")

def delete_gorev():
    show_gorev()
    try:
        silinecek = int(input("Silmek istediğiniz görevin numarasını girin: "))
        if 1 <= silinecek <= len(gorev_list):
            silinen = gorev_list.pop(silinecek - 1)
            print(f"'{silinen}' silindi.")
        else:
            print("Geçersiz numara!")
    except ValueError:
        print("Lütfen bir sayı girin.")

def exit_program():
    print("Programdan çıkılıyor... Görüşmek üzere!")
    return True

while True:
    print("\nYapılacaklar Listesi")
    print("1. Görev Ekle")
    print("2. Görevleri Göster")
    print("3. Görev Sil")
    print("4. Çık")

    secim = input("Seçiminiz (1-4): ")

    if secim == "1":
        add_gorev()
    elif secim == "2":
        show_gorev()
    elif secim == "3":
        delete_gorev()
    elif secim == "4":
        if exit_program():
            break
    else:
        print("Geçersiz seçim. Lütfen 1-4 arasında bir sayı girin.")
```
