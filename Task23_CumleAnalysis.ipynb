```python
from google.colab import files
import re
import string

# Dosyayı yükle
uploaded = files.upload()
dosya_adi = list(uploaded.keys())[0]

# Dosyayı oku
with open(dosya_adi, "r", encoding="utf-8") as dosya:
    text = dosya.read().lower()  # Küçük harfe çevir

# Cümlelere ayır
cumleler = re.split(r"[.!?]", text)
cumleler = [c.strip() for c in cumleler if c.strip() != ""]

# Harf sıklığı her cümle için
for idx, cumle in enumerate(cumleler, 1):
    temiz = re.sub(r"[^a-z]", "", cumle)  # Sadece harfleri bırak
    harf_sayisi = {}
    for harf in string.ascii_lowercase:
        harf_sayisi[harf] = temiz.count(harf)

    print(f"\n🟩 Cümle {idx}: {cumle}")
    print(harf_sayisi)
```
